```python
s = data.iloc[1:, [4, 7, 10, 13, 16]]
fig, ax = plt.subplots(figsize=(18, 12))
y = np.array([i for i in range(1, 37)])

w = 0.15
x = np.array(s.iloc[:, 0], dtype=float)
ax.bar(y - 2 * w , x, width=w,color='b',align='center')

x = np.array(s.iloc[:, 1], dtype=float)
ax.bar(y - w,   x, width=w,color='g',align='center')

x = np.array(s.iloc[:, 2], dtype=float)
ax.bar(y, x, width=w,color='r',align='center')

x = np.array(s.iloc[:, 3], dtype=float)
ax.bar(y + w , x, width=w,color='y',align='center')

x = np.array(s.iloc[:, 4], dtype=float)
ax.bar(y + 2 * w , x, width=w,color='purple',align='center')

plt.legend(['Харрман', 'Фано', 'LZ77_4', 'LZ77_8', 'LZ77_16'])
plt.title("Среднее время распаковки")
plt.ylabel("Среднее время распаковки")
plt.xlabel("N файла")
```
